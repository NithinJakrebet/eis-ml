```python
# Standard Imports
import sys
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ensure the scripts directory is in sys.path
sys.path.append(os.path.abspath("../scripts"))

# Import the custom load function
from load_data import load_data
```

```python
# Load battery impedance dataset
data = load_data()

# Show available datasets (file names)
print("Available datasets:", data.keys())
```

```python
# Choose one dataset to analyze (adjust filename as needed)
df_a1 = data["Channel A1.csv"]

# Display first few rows
df_a1.head()
```

```python
# Iterate through all datasets and display the first few rows
for filename, df in data.items():
    print(f"\nDataset: {filename} (Shape: {df.shape})")
    display(df.head())  # Jupyter-friendly display
```

```python
# Select only datasets where Ns is 1 or 6, and cycle number is not 0
filtered_df = df_a1.loc[(df_a1['Ns'].isin([1, 6])) & (df_a1['cycle number'] != 0)].copy()

# Extract Real and Imaginary impedance values
Re_Z = filtered_df['Re(Z)/Ohm'].values
Im_Z = filtered_df['Im(Z)/Ohm'].values
```

```python
# Create Nyquist plot
plt.figure(figsize=(10, 10))
plt.plot(Re_Z, Im_Z, 'o', markersize=5, alpha=0.7, label="Impedance Data")

# Labels and formatting
plt.xlabel('Re(Z) / Ohm')
plt.ylabel('-Im(Z) / Ohm')
plt.title('Nyquist Plot of Battery Impedance')
plt.grid(True)
plt.axis('equal')
plt.legend()
```

```python
# Save the plot before showing it
plt.savefig("../results/nyquist_plot_A1.png", dpi=300, bbox_inches="tight")
```
